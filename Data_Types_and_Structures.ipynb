#Data Types and Structures



1. What are data structures, and why are they important?
Data structures are organized ways to store and manage data so operations like access, insertion, and deletion can be done efficiently. They are important because they improve performance, memory usage, and problem-solving efficiency.

2. Explain the difference between mutable and immutable data types with examples.
Mutable types can be changed after creation (examples: list, dict, set).
Immutable types cannot be changed after creation (examples: int, float, str, tuple).

3. What are the main differences between lists and tuples in Python?
Lists are mutable, use square brackets, and support modification.
Tuples are immutable, use parentheses, and represent fixed data.

4. Describe how dictionaries store data.
Dictionaries store key-value pairs using a hash table. A hash function computes the index where each value is stored.

5. Why might you use a set instead of a list in Python?
Sets remove duplicates automatically and provide faster membership testing than lists.

6. What is a string in Python, and how is it different from a list?
A string is an immutable sequence of characters. A list is a mutable sequence that can contain any type of elements.

7. How do tuples ensure data integrity in Python?
Tuples are immutable, so their values cannot be changed after creation. This prevents accidental modifications.

8. What is a hash table, and how does it relate to dictionaries in Python?
A hash table stores key-value pairs using a hash function. Python dictionaries are implemented using a hash table.

9. Can lists contain different data types in Python?
Yes, lists can contain multiple data types (heterogeneous data).

10. Explain why strings are immutable in Python.
Strings are immutable for safety, memory optimization, and to allow them to be used as dictionary keys.

11. What advantages do dictionaries offer over lists for certain tasks?
Dictionaries allow fast key-based lookups (O(1) on average) and are better for mapping identifiers to values.

12. Describe a scenario where using a tuple would be preferable over a list.
When storing constant data like coordinates, RGB values, or configuration records that should not change.

13. How do sets handle duplicate values in Python?
Sets automatically remove duplicates when elements are added.

14. How does the “in” keyword work differently for lists and dictionaries?
For lists, it checks if a value exists in the list.
For dictionaries, it checks if a key exists in the dictionary.

15. Can you modify the elements of a tuple? Explain why or why not.
No, you cannot modify tuple elements because tuples are immutable.

16. What is a nested dictionary, and give an example of its use case.
A nested dictionary is a dictionary inside another dictionary.
Example use case: storing user data with attributes for each user.

Example:

users = {
    "u1": {"name": "Alice", "age": 30},
    "u2": {"name": "Bob", "age": 25}
}


17. Describe the time complexity of accessing elements in a dictionary.
Accessing a value by key is O(1) on average.

18. In what situations are lists preferred over dictionaries?
Lists are preferred when order matters, when indexing is required, or when storing simple sequences of items.

19. Why are dictionaries considered unordered, and how does that affect data retrieval?
Dictionaries conceptually map keys to values without positional order. Retrieval is based on keys, not index positions.

20. Explain the difference between a list and a dictionary in terms of data retrieval.
A list retrieves data by index (position).
A dictionary retrieves data by key.
